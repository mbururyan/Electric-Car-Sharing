```python
# Most popular station

#Choosing only stations
df_station = df[df['Kind'] == 'STATION']
df_station

#Group the stations by count to find what station occurs most times

df_grouped = df_station.groupby('ID').count()
df_grouped

#Sorting the grouped stations

pop_stations = df_grouped.sort_values(by='Cars', ascending=0)
pop_stations.head()
```

```python
# Crucial Data Cleaning

# Separating datetime columns into date and time
df.head()

#date column
df['date'] = pd.to_datetime((df.year*10000+df.month*100+df.day).apply(str),format='%Y%m%d')
df.head(1)

```

```python
# Changing new_datetime to datetime datatype
df['new_datetime'] = pd.to_datetime(df['new_datetime'])
df.info()
```

```python
#testing the time col

df['new_datetime'].dt.hour
```

```python
# Most popular hour for picking blue cars

#groupby hour
grp_hour_blue_pick = df.groupby('hour')[['Bluecar counter']].sum()
grp_hour_blue_pick

#sort hours by bluecars
pop_hour_blue = grp_hour_blue_pick.sort_values(by='Bluecar counter', ascending=0)
pop_hour_blue.head()
```

#### Most popular hour for bluecars is 6AM

```python
# Least popular hour for bluecars
least_pop_hour_blue = grp_hour_blue_pick.sort_values(by='Bluecar counter', ascending=1)
least_pop_hour_blue.head()
```

#### Least popular hour for blue cars is 1900 HRS / 9 PM

```python
# Popular hour for Utilib cars

#Grouped per hour
grp_hour_utilib = df.groupby('hour')[['Utilib counter']].sum()
grp_hour_utilib

#Ranked from most popular
pop_hour_utilib = grp_hour_utilib.sort_values(by='Utilib counter', ascending=0)
pop_hour_blue.head()
```

#### Most popular hour for Utilib cars is 6 AM 

```python
# Least popular hour for Utilib cars
least_pop_hour_utilib = grp_hour_utilib.sort_values(by='Utilib counter', ascending=1)
least_pop_hour_utilib.head()
```

#### Least popular hour when a Utilib vehicle was picked is 2300 HRS / 11 PM

```python
# Most popular hour for utilib 1.4 

group_utilib_4 = df.groupby('hour')[['Utilib 1.4 counter']].sum()
group_utilib_4

pop_hour_utilib_4 = group_utilib_4.sort_values(by='Utilib 1.4 counter', ascending=0)
pop_hour_utilib_4.head()
```

#### Most popular hour for utilib 1.4 vehicles is 3 AM

```python
# least popular hour for utilib 1.4
least_pop_hour_utilib_4 = group_utilib_4.sort_values(by='Utilib 1.4 counter', ascending=1)
least_pop_hour_utilib_4.head()
```

#### Least popular hour for utilib 1.4 is 1700 HRS / 5 PM

```python
# Most popular station during popular picking hour

# Most popular hour is 8 AM
pop_hour = df.loc[df.hour == 8]
pop_hour

pop_hour.groupby('ID')['Bluecar counter'].sum().sort_values(ascending=False).head()
```
